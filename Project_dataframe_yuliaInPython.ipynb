#TASK OF THE MINI PROJECT

**1. Nilai rata-rata akhir (G3) semua murid di sekolah GP**

```python
df["school"] == "GP"
```

```python
#subsetting to display suitable data
anak_gp = df[df["school"] == "GP"]
anak_gp
```

```python
anak_gp_mean = anak_gp["G3"].mean()
anak_gp_mean = round(anak_gp_mean, 2)
print("Nilai rata-rata akhir (G3) semua murid di sekolah 'GP':", anak_gp_mean)
```

**2. Nilai min dan max (G3) dari murid di sekolah "MS"**

```python
g3_ms = df[df["school"] == "MS"]["G3"]
min_g3_ms = g3_ms.min()
max_g3_ms = g3_ms.max()

print("Nilai terendah dari murid di sekolah MS:", min_g3_ms)
print("Nilai tertinggi dari murid di sekolah MS:", max_g3_ms)
```

**7. Jumlah anak yang meningkat skor ujiannya (first vs. final)**
